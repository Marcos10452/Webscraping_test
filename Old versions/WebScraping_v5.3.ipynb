```python
#Function to detect whether or not the excel file was created
def CreateFile(CFdf,ChangeDateFile,sheetname):
    w_filenameXLSX='../../Python_ITBA/My test/ExcelFile/Web-Scraping_V5.1_'+ChangeDateFile+'.xlsx'
    print ("File exists:"+str(path.exists(w_filenameXLSX)))
    if  not(path.exists(w_filenameXLSX)):
        # Create a Pandas Excel writer using XlsxWriter as the engine.
        writer = pd.ExcelWriter(w_filenameXLSX, engine='xlsxwriter')
    else:
        writer = pd.ExcelWriter(w_filenameXLSX, engine='openpyxl')
        # try to open an existing workbook
        writer.book = load_workbook(w_filenameXLSX)
        # copy existing sheets
        writer.sheets = dict((ws.title, ws) for ws in writer.book.worksheets)
        # read existing file
        reader = pd.read_excel(w_filenameXLSX)
        # write out the new sheet
    # Close the Pandas Excel writer and output the Excel file.
    CFdf.to_excel(writer,sheet_name=sheetname,index=False)
    print("saved:"+sheetname)
    writer.save()
```

```python
#Function to generate URL
def URLgenerate(DealerName,AuxText,URLInd,IndexNumber):
    
    if  (DealerName=='claro'):
        if (URLInd=='URL1'):
            return("https://tienda.claro.com.ar/#facet:&productBeginIndex:"+str(IndexNumber*12)+"&orderBy:6&pageView:grid&minPrice:&maxPrice:&pageSize:&")
        else:
            return("https://tienda.claro.com.ar/#facet:&productBeginIndex:"+str(IndexNumber*12)+"&orderBy:6&pageView:grid&minPrice:&maxPrice:&pageSize:&")
    
    elif  DealerName=='movistar':
        if (URLInd=='URL1'):
            return("https://tienda.movistar.com.ar/?adgroupid=58700004606889709&aux_utm_campaign=GS+-+Brand+-+Tienda+Movistar+-+72460&ds_e_adid=330789885539&ds_e_device=c&ds_e_matchtype=search&ds_e_network=g&ds_s_kwgid=58700004606889709&ds_url_v=2&gclid=EAIaIQobChMI3KGg7e_Z6AIVjYORCh0tVgZMEAAYASAAEgLmLvD_BwE&gclsrc=aw.ds&keyword=p38771900175&lid=43700038771900175&p="+str(IndexNumber+1))
        else:
            return("https://tienda.movistar.com.ar/?adgroupid=58700004606889709&aux_utm_campaign=GS+-+Brand+-+Tienda+Movistar+-+72460&ds_e_adid=330789885539&ds_e_device=c&ds_e_matchtype=search&ds_e_network=g&ds_s_kwgid=58700004606889709&ds_url_v=2&gclid=EAIaIQobChMI3KGg7e_Z6AIVjYORCh0tVgZMEAAYASAAEgLmLvD_BwE&gclsrc=aw.ds&keyword=p38771900175&lid=43700038771900175&p="+str(IndexNumber+1))
   
    elif  DealerName=='personal':
        if (URLInd=='URL1'):
            return("https://tienda.personal.com.ar/equipos")
        else:
            return("https://tienda.personal.com.ar/equipos/?PageNumber="+str((IndexNumber)+1))
    elif  DealerName=='castillo':
        if (URLInd=='URL1'):
            return("https://www.castillo.com.ar/"+AuxText)
        else:
            return("https://www.castillo.com.ar/"+AuxText+"_Desde_"+str((IndexNumber*50)+1))
    elif  (DealerName=='Cetrogar'):
        if (URLInd=='URL1'):
            return("https://www.cetrogar.com.ar/"+AuxText+str(1+IndexNumber))
        else:
            return("https://www.cetrogar.com.ar/"+AuxText+str(1+IndexNumber))
    elif  DealerName=='Fravega':
        if (URLInd=='URL1'):
            return("https://www.fravega.com/l/?categorias="+AuxText+"&page="+str(IndexNumber+1))
        else:
            return("https://www.fravega.com/l/?categorias="+AuxText+"&page="+str(IndexNumber+1))
    elif  DealerName=='Musimundo':
        if (URLInd=='URL1'):
            return("https://www.musimundo.com/"+AuxText+"Arelevance&page="+str(IndexNumber+1))
        else:
            return("https://www.musimundo.com/"+AuxText+"Arelevance&page="+str(IndexNumber+1))
```

```python
for SuperArray in range(len(Array)):
    CategoryType=Array[SuperArray][1]


    ExcelFile=Array[SuperArray][2]

    FindPages=Array[SuperArray][3]

    ClassFindPager=Array[SuperArray][4]

    CountPages=Array[SuperArray][5] # count how much pages are


    #Finding the product Level 0
    L0_FindType=Array[SuperArray][6]
    L0_FindClass=Array[SuperArray][7]

    #Getting Name
    FindNameType=Array[SuperArray][8]
    FindNameClass=Array[SuperArray][9]
    
    #Finding price Level 1
    L1_FindType=Array[SuperArray][10]
    L1_FindClass=Array[SuperArray][11]
    
    #Finding price Level 2
    L2_FindType=Array[SuperArray][12]
    L2_FindClass=Array[SuperArray][13]
    
     #Getting price list 
    FindPriceListType=Array[SuperArray][14]
    FindPriceListClass=Array[SuperArray][15]
    
    #Getting Final Price
    FindPriceType=Array[SuperArray][16]
    FindPriceClass=Array[SuperArray][17]
    
    #Getting Regular Price
    FindRegularPriceType=Array[SuperArray][18]
    FindRegularPriceClass=Array[SuperArray][19]
    
    #Getting discount number
    FindPriceOffType=Array[SuperArray][20]
    FindPriceOffClass=Array[SuperArray][21]
    
    
    #Define today
    x = datetime.datetime.now()
    TodayDate=x.strftime("%Y"+'-'+"%m"+'-'+"%d")


    # Active Chrome with chromedriver
    #Check vulnerability in port 9515 due to selenium
    driver = webdriver.Chrome("C:\ChromeDriver\chromedriver.exe")
    
    NowDate=[]
    Category=[]
    Brand=[]
    products=[] #List to store name of the product
    PriceList=[]
    Discount=[]
    prices=[] #List to store price of the product
       
    Position=[] #List of position
    PageN=[] #List of pages
    
    #1st stage, looking for brands
    for NextBrand in CategoryType:

        UrlAux=URLgenerate(Array[SuperArray][0],NextBrand,'URL1',0) 
        
        driver.get(UrlAux)
        time.sleep(1)
        content = driver.page_source
        time.sleep(1)
        soup = BeautifulSoup(content)
        time.sleep(1)

       # print(soup.find('ul',attrs={'class':'andes-pagination'}).prettify())
        TotalNum=-1
        print(bool(soup.find(FindPages,attrs={'class':ClassFindPager})))
        if bool(soup.find(FindPages,attrs={'class':ClassFindPager}))==True:
            PagNum=soup.find(FindPages,attrs={'class':ClassFindPager})
            #print(PagNum)
            #TotalNum=len(PagNum.findAll(CountPages))-1
            TotalNum=len(PagNum.findAll(CountPages))-1
            print(CountPages)

        #TotalNum has the total number of how many pages are. every page has 48 elements
        #If TotalNum=0 means that all products are in one page only.
        if TotalNum<0:
            TotalNum=1
        
        TotalNum=10
        print(TotalNum)
        #2nd stage, to browse in all pages more then 48 products each.
        for i in  range(TotalNum):
            print(i)
            if i>0 :
                # Creating the URL based on informartion from web page
                print('URL2')
                UrlAux=URLgenerate(Array[SuperArray][0],NextBrand,'URL2',i) 
                time.sleep(1)
                driver.get(UrlAux)
                time.sleep(1)
                content = driver.page_source
                time.sleep(1)
                soup = BeautifulSoup(content)
                time.sleep(1)
           

            print(UrlAux)
            #print("div: "+str(FindAllType))
            #print("class: "+str(FindAllClass))
            #print("Nombre: "+str(soup.findAll(FindAllType,attrs={'class':FindAllClass})))
            #print( soup.findAll('section',attrs={'class':'results grid'}))
            #3rd  stage, looking for name and price
            idPostion=0
            for Level0 in soup.findAll(L0_FindType,attrs={'class':L0_FindClass}):
                name=Level0.find(FindNameType, attrs={'class':FindNameClass})
                #print("Datos ")
                if bool(name)==True:
                    NowDate.append(TodayDate)
                    Category.append(CategoryType[NextBrand])
                    #Getting Brand
                    BoolTest=False
                    for word in name.text.replace("\n", "").split():
                        for BrandCheck in DictBrands:
                            if ((word.upper())==BrandCheck) and (BoolTest==False):
                                Brand.append(DictBrands[word.upper()])
                                BoolTest=True
                    if BoolTest==False:
                        Brand.append(' ')
                            
                    products.append(name.text.replace("\n", ""))
                    
                    #Level 1
                    Level1=Level0.find(L1_FindType, attrs={'class':L1_FindClass})
                    #Level 2
                    Level2=Level1.find(L2_FindType, attrs={'class':L2_FindClass})
                    
                    #Price List
                    if bool(Level2.find(FindPriceListType, attrs={'class':FindPriceListClass}))==True:
                        AuxPriceList=Level2.find(FindPriceListType, attrs={'class':FindPriceListClass})
                        PriceList.append((AuxPriceList.text.replace("\n", "")).replace(" ",""))
                    #if the price list is the only one
                    elif bool(Level2.find(FindRegularPriceType, attrs={'class':FindRegularPriceClass}))==True:
                        AuxPriceList=Level2.find(FindRegularPriceType, attrs={'class':FindRegularPriceClass})
                        PriceList.append((AuxPriceList.text.replace("\n", "")).replace(" ",""))
                    else:
                        PriceList.append(0)

                    #Final Price   
                    if bool(Level2.find(FindPriceType, attrs={'class': FindPriceClass}))==True:
                        price=Level2.find(FindPriceType, attrs={'class':FindPriceClass})
                        prices.append((price.text.replace("\n", "")).replace(" ",""))
                    else:
                        prices.append(0)

                    #Discount . If it was not found
                    #Discount . Queda afuera porque esta distribuido en diferentes luegars
                    if (bool(Level1.find(FindPriceOffType, attrs={'class':FindPriceOffClass}))==True):
                        Discount.append(Level1.find(FindPriceOffType, attrs={'class':FindPriceOffClass}).text)
                    elif (bool(Level2.find(FindPriceOffType, attrs={'class':FindPriceOffClass}))==True):
                        Discount.append(Level2.find(FindPriceOffType, attrs={'class':FindPriceOffClass}).text)
                    else:
                        Discount.append(0)
                            
                    idPostion+=1
                    Position.append(idPostion)
                    PageN.append(i+1)
                    
                    #print(products)
            #print(products)
            #driver.quit()
            
        df=pd.DataFrame({'Date':NowDate,'Category':Category,'Brand':Brand,'Product Name':products,'Price List':PriceList,
                    'Price Off':Discount,'Price':prices,'Position':Position,'PageNumber':PageN}) 
    
        #Saving file as dealer and adding the reporting date
        CreateFile(df,str(TodayDate),(Array[SuperArray][0]))
       

    driver.quit()
```
