```python
#Selenium is a web testing library. It is used to automate browser activities.
from selenium import webdriver
#Beautiful Soup is a Python package for parsing HTML and XML documents. 
#It creates parse trees that is helpful to extract the data easily.
from bs4 import BeautifulSoup
import pandas as pd
from openpyxl import load_workbook
#Manage files and directories
import os
from os import path
```

```python
#Function to detect whether or not the excel file was created
def CreateFile(CFdf,sheetname):
    print ("File exists:"+str(path.exists('Cetrogar.xlsx')))
    if  not(path.exists('Cetrogar.xlsx')):
        # Create a Pandas Excel writer using XlsxWriter as the engine.
        writer = pd.ExcelWriter('Cetrogar.xlsx', engine='xlsxwriter')
    else:
        writer = pd.ExcelWriter('Cetrogar.xlsx', engine='openpyxl')
        # try to open an existing workbook
        writer.book = load_workbook('Cetrogar.xlsx')
        # copy existing sheets
        writer.sheets = dict((ws.title, ws) for ws in writer.book.worksheets)
        # read existing file
        reader = pd.read_excel('Cetrogar.xlsx')
        # write out the new sheet
        
    # Close the Pandas Excel writer and output the Excel file.
    CFdf.to_excel(writer,sheet_name=sheetname,index=False)
    print("saved:"+sheetname)
    writer.save()
```

# Active Chrome with chromedriver

```python
#Check vulnerability in port 9515 due to selenium
```

```python
driver = webdriver.Chrome("C:\ChromeDriver\chromedriver.exe")
```

```python
# selecting brands
```

```python
brands=['tecnologia/celulares-accesorios','tecnologia/tv-video',
        'electrodomesticos/heladeras-y-freezers']
```

```python
#1st stage, looking for brands
for NextBrand in brands:
    
    products=[] #List to store name of the product
    prices=[] #List to store price of the product
    UrlAux="https://www.cetrogar.com.ar/"+NextBrand+".html"
    driver.get(UrlAux)
    content = driver.page_source
    soup = BeautifulSoup(content)
    
   # print(soup.find('ul',attrs={'class':'andes-pagination'}).prettify())
    TotalNum=-1
    if bool(soup.find('div',attrs={'class':'pager'}))==True:
        PagNum=soup.find('div',attrs={'class':'pager'})
        TotalNum=len(PagNum.findAll('li'))-1

    #TotalNum has the total number of how many pages are. every page has 48 elements
    #If TotalNum=0 means that all products are in one page only.
    if TotalNum<0:
        TotalNum=1
    print(TotalNum)
    #2nd stage, to browse in all pages more then 48 products each.
    for i in  range(TotalNum):
        if i>0 :
            # Creating the URL based on informartion from web page
            UrlAux="https://www.cetrogar.com.ar/"+NextBrand+".html?p="+str(i+1)
            driver.get(UrlAux)
            content = driver.page_source
            soup = BeautifulSoup(content)

        print(UrlAux)
        #print( soup.findAll('section',attrs={'class':'results grid'}))
        #3rd  stage, looking for name and price
        for a in soup.findAll('li',attrs={'class':['item even','item odd']}):
            name=a.find('h2', attrs={'class':'product-name'})
            price=a.find('span', attrs={'class':'price'})
            products.append(name.text)
            #Replace tab for nothing
            prices.append(price.text.replace("\t", ""))
        #print(products)
    df=pd.DataFrame({'Product Name':products,'Price':prices}) 

    CreateFile(df,(NextBrand.split('/')[1]))

driver.quit()
```
